#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python

#use Z values -> observed value - mean / sd
#create a new df of normalised prices
def standard_normalisation(column):
    mean = column.mean()
    std = column.std()
    return (column - mean)/std

#from tqdm import tqdm

normalised_df = grouped['price'].apply(standard_normalisation)
print(normalised_df)

plt.rcParams['figure.figsize'] = (12.0, 10.0) #resize plots
plt.hist(normalised_df, bins=30, edgecolor='black')
plt.show()



```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
"""
x = south_belt_df['surface']
y = south_belt_df['price']
plt.scatter(x, y)
plt.title('Grouped scatter plots of price vs surface area of houses in Madrid')
plt.show()
"""
#JSeiglie's code below
populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]  


fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Scatter Plots of Surface Area vs. Price by Population", fontsize=16)

for i, population in enumerate(populations):
    row = i // 2
    col = i % 2
    ax = axs[row, col]
    
    
    population_data = ds[ds['level5'] == population]
    
 
    ax.scatter(population_data['surface'], population_data['price'], alpha=0.5)
    ax.set_title(f"Population: {population}")
    ax.set_xlabel("Surface Area (sqm)")
    ax.set_ylabel("Price")
    ax.grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
